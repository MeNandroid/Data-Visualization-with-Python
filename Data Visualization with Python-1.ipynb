```python
import numpy as np
import pandas as pd
```

```python
data ="https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DV0101EN/labs/coursera/Topic_Survey_Assignment.csv"
```

```python
df=pd.read_csv(data,index_col = 0)
print ('Done reading into a pandas dataframe!')
```

```python
#display data
df.head()
```

```python
#display info
df.info
```

```python
# determine version
%matplotlib inline 

import matplotlib as mpl
import matplotlib.pyplot as plt

mpl.style.use('ggplot') # optional: for ggplot-like style

# check for latest version of Matplotlib
print ('Matplotlib version: ', mpl.__version__)
```

```python
#sort and display
df.sort_values(['Very interested'], ascending=False, axis=0, inplace=True)

# Taking the percentage of the responses and rounding it to 2 decimal places 
df= round((df/2233)*100,2)

# View top 5 rows of the data 
df.head()
```

```python
# make a plot and show
ax = df.plot(kind='bar', 
                figsize=(20, 8),
                rot=90,color = ['#5cb85c','#5bc0de','#d9534f'],
                width=.8,
                fontsize=14)

ax.set_title('Percentage of Respondents Interest in Data Science Areas',fontsize = 16)

ax.set_facecolor('white')

ax.legend(fontsize=14,facecolor = 'white') 

ax.get_yaxis().set_visible(False)

for p in ax.patches:
    ax.annotate(np.round(p.get_height(),decimals=2), 
                (p.get_x()+p.get_width()/2., p.get_height()), 
                ha='center', 
                va='center', 
                xytext=(0, 10), 
                textcoords='offset points',
                fontsize=14
               )
plt.show()
```
